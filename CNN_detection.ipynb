```python
import pandas as pd
import numpy as np
import os
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler


from google.colab import drive
drive.mount('/content/drive')

data_dir = '/content/drive/MyDrive/Independent Study/aptos2019-blindness-detection'
save_dir = '/content/drive/MyDrive/Independent Study'

train_df = pd.read_csv(f'{data_dir}/train.csv')
train_df['id_code'] = train_df['id_code'].astype(str) + '.png'
print("Total training samples:", len(train_df))
print("Class distribution:")
print(train_df['diagnosis'].value_counts().sort_index())
print()

from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1./255)

train_gen = datagen.flow_from_dataframe(
    train_df,
    directory=f'{data_dir}/train_images',
    x_col='id_code',
    y_col='diagnosis',
    class_mode='raw',
    target_size=(224,224),
    batch_size=32,
    shuffle=False
)

from tensorflow.keras.applications import EfficientNetB0

base_model = EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3),
    pooling='avg'
)

base_model.trainable = False

print("Model Architecture:")
print(f"Base Model: EfficientNetB0 (pretrained on ImageNet)")
print(f"Pooling: Global Average Pooling")
print(f"Total Parameters: {base_model.count_params():,}")
print()

print("Extracting features from training images...")
steps = math.ceil(len(train_df) / 32)
features = base_model.predict(train_gen, steps=steps, verbose=1)

print("\n Feature Extraction Complete")
print(f"Extracted features shape: {features.shape}")
print(f"Number of images: {features.shape[0]}")
print(f"Feature vector dimensionality: {features.shape[1]}")
print()

np.save(f'{save_dir}/train_features.npy', features)
np.save(f'{save_dir}/train_labels.npy', train_df['diagnosis'].values)

print("Saved:")
print(f"  - {save_dir}/train_features.npy")
print(f"  - {save_dir}/train_labels.npy")
print()

print("="*70)
print("FEATURE EXTRACTION ANALYSIS")
print("="*70)
print()

feature_mean = features.mean(axis=0)
feature_std = features.std(axis=0)
feature_min = features.min(axis=0)
feature_max = features.max(axis=0)

print("Feature Vector Statistics :")
print(f"  Mean value: {feature_mean.mean():.6f}")
print(f"  Std deviation: {feature_std.mean():.6f}")
print(f"  Min value: {feature_min.min():.6f}")
print(f"  Max value: {feature_max.max():.6f}")
print()

labels = train_df['diagnosis'].values
print("Feature Statistics by Diabetic Retinopathy Severity Class:")
print()

class_stats = []
for class_idx in sorted(np.unique(labels)):
    class_mask = labels == class_idx
    class_features = features[class_mask]

    stats_dict = {
        'Class': class_idx,
        'Class_Name': ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative DR'][class_idx],
        'Sample_Count': class_mask.sum(),
        'Feature_Mean': class_features.mean(),
        'Feature_Std': class_features.std(),
        'Feature_Min': class_features.min(),
        'Feature_Max': class_features.max()
    }
    class_stats.append(stats_dict)

    print(f"Class {class_idx} - {stats_dict['Class_Name']}:")
    print(f"  Samples: {stats_dict['Sample_Count']}")
    print(f"  Mean: {stats_dict['Feature_Mean']:.6f}, Std: {stats_dict['Feature_Std']:.6f}")
    print()

class_stats_df = pd.DataFrame(class_stats)
print()


print("Displaying Class-wise Feature Statistics:")
display(class_stats_df)
print()

print("First 5 Feature Vectors:")
feature_df = pd.DataFrame(
    features[:5],
    columns=[f'Feature_{i}' for i in range(features.shape[1])]
)
display(feature_df)
print()


plt.figure(figsize=(16, 2))
sns.heatmap([features[0]], cmap="viridis", cbar_kws={'label': 'Activation Value'})
plt.title("Feature Vector Heatmap - Image 0\n(1280 features extracted from EfficientNetB0)",
          fontsize=12, fontweight='bold')
plt.xlabel("Feature Index")
plt.ylabel("Sample ID")
plt.tight_layout()
plt.savefig(f'{save_dir}/feature_single_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
print()

plt.figure(figsize=(16, 4))
sns.heatmap(features[:10], cmap="magma", cbar_kws={'label': 'Activation Value'})
plt.title("Feature Vectors Heatmap - First 10 Images\n(Each row = 1280-dimensional feature vector)",
          fontsize=12, fontweight='bold')
plt.xlabel("Feature Index")
plt.ylabel("Sample Index")
plt.tight_layout()
plt.savefig(f'{save_dir}/feature_10samples_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
print()


plt.figure(figsize=(14, 5))
for class_idx in sorted(np.unique(labels)):
    class_mask = labels == class_idx
    class_name = ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative DR'][class_idx]
    class_features_mean = features[class_mask].mean()
    plt.hist(features[class_mask].flatten(), bins=50, alpha=0.6, label=f'Class {class_idx}: {class_name}')

plt.xlabel("Feature Activation Value")
plt.ylabel("Frequency")
plt.title("Distribution of Feature Values Across DR Severity Classes", fontsize=12, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{save_dir}/feature_distribution_by_class.png', dpi=300, bbox_inches='tight')
plt.show()
print()

print("Computing t-SNE visualization")

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

tsne = TSNE(n_components=2, random_state=42, n_iter=1000, perplexity=30)
features_tsne = tsne.fit_transform(features_scaled)

plt.figure(figsize=(12, 8))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
class_names = ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative DR']

for class_idx in sorted(np.unique(labels)):
    class_mask = labels == class_idx
    plt.scatter(
        features_tsne[class_mask, 0],
        features_tsne[class_mask, 1],
        c=colors[class_idx],
        label=f'Class {class_idx}: {class_names[class_idx]}',
        alpha=0.7,
        s=50
    )

plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.title("t-SNE Visualization of Extracted Features\n(Shows how well different DR severity levels are separated in feature space)",
          fontsize=12, fontweight='bold')
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{save_dir}/feature_tsne_visualization.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ t-SNE visualization complete")
print()

print("Computing feature correlation matrix using 100 features")

feature_subset_indices = np.arange(0, features.shape[1], 13)
features_subset = features[:, feature_subset_indices]

corr_matrix = np.corrcoef(features_subset.T)

plt.figure(figsize=(10, 9))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0, square=True,
            cbar_kws={'label': 'Correlation Coefficient'})
plt.title("Feature Correlation Matrix (100 sampled features)\n(Shows dependencies between extracted features)",
          fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{save_dir}/feature_correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
print()


plt.figure(figsize=(14, 5))

class_feature_means = []
for class_idx in sorted(np.unique(labels)):
    class_mask = labels == class_idx
    class_feature_means.append(features[class_mask].mean(axis=0))

class_feature_means = np.array(class_feature_means)

for class_idx in range(len(class_feature_means)):
    plt.plot(class_feature_means[class_idx], alpha=0.7,
             label=f'Class {class_idx}: {class_names[class_idx]}', linewidth=1.5)

plt.xlabel("Feature Index")
plt.ylabel("Mean Activation Value")
plt.title("Average Feature Activation by DR Severity Class\n(Shows how different pathologies activate different features)",
          fontsize=12, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{save_dir}/feature_class_means.png', dpi=300, bbox_inches='tight')
plt.show()
print()
```
